### **Definition**
Loops in Python are control flow statements that allow you to repeat a block of code multiple times.  They are essential for automating repetitive tasks, such as processing items in a list, performing calculations, or reading data from files.


### **For loop**
used to iterates over a sequence(like a list, tuple, string, or range) or iterable object.
```python
for i in range(a,b,c):
    print(i)
```    

here a is starting point, b is end point, and c is skip steps

````python
# range : if range(10) then it will print 0 to 9 it means it will print 10 numbers but it will start from 0 to 9
for i in range(10):
    print(i)

for x in range(1,11): # if range(1,11) then it will print 1 to 10 it means it will print 10 numbers but it will start from 1 to 10
    print(x)

for y in range(1,11,2): # if range(1,11,2) then it will print 1 to 10 it means it will print 5 numbers but it will start from 1 to 10 with step of 2
    print(y)

for a in range(10,0,-1): # if range(10,0,-1) then it will print 10 to 1 it means it will print 10 numbers but it will start from 10 to 1 with step of -1
    print(a)   

str = "Hello World"
for s in str:
    print(s)          
````

### **While loop**
used to continue executing as a long as a specified condition remains ```true``` making them ideal when the number of iterations is not known in advance.

```python
x = 0
while x>6:
    print(x)
    x+=1
```    

````python
count = 0
while count<5:
    print(count)
    count+=1 
````

### **Loop Control Statements**
Loop Control Statements are break, continue and pass statements which are used to control the flow of the loop.

```break``` : exits the loop,
```continue``` : skip to the next iteration,
```pass``` : acts as a placeholder.


````python
#break and continue and pass statement in loops
for i in range(10):
    if i==5:
        break
    print(i)

for i in range(10):
    if i==5:
        continue
    print(i)

for i in range(10):
    if i ==5:
        pass
    print(i)    
````

### **Nested Loops**
Nested loops in Python involve placing one loop inside another, enabling iteration over multiple sequences or performing repeated actions with multiple layers of repetition.


````python
for i in range(11):
    for j in range(11):
        print(f"{i} x {j} = {i*j}")      
````

````python
## Examples:
#calculate the sum of first n natural numbers using a while and for loop

#by for loop
n = int(input("Enter your number: "))
sum = 0
for i in range(n):
    sum+=i
print(f"the sum of first {n} natural numbers is {sum}")

#by while loop
m = int(input("Enter your number: "))
sum = 0
count = 0
while count<m:
    sum+=count
    count+=1
print(f"the sum of first {m} natural numbers is {sum}") 
````

````python
## Prime numbers between 1 to n
n = int(input("Enter your number: "))
for i in range(2,n+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i)
        
````

### **Conclusion**
Loops are powerful constructs in Python that allow you to execute a block of code multiple times. By understanding and using for and while loops, along with loop control statements like break, continue, and pass, you can handle a wide range of programming tasks efficiently.
