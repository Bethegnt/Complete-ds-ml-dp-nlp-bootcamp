```python
## if statement
age = 25
if age>=18:
    print("You are allowed to vote in the elections")    
```

```python
## else
# the else statement execute a block of code if the condition in the if statement is false
age = 16
if age>18:
    print("You are eligible for voting")
else:
    print("You are a minor")     
```
