### **Definition**
**Conditional Statements** are used to control the flow of a program based on certain conditions. They allow you to execute different blocks of code depending on whether a condition is true or false. In Python, there are several types of conditional statements, including if statements, if elif statements, if else statements, and nested if statements.


### **if Statement**
if statements are used to check a condition and execute a block of code if the condition is true. The syntax for an if statement is as follows:
 ```python
 if condition:
     code to execute if condition is true
```     

````python
age = 25
if age>=18:
    print("You are allowed to vote in the elections")    
````

### **else statement**
else statements are used to execute block of code if a condition is true and another block of code if the condition is false. The syntax for an if else statement is as follows:
 ```python
 if condition:
      code to execute if condition is true
 else:
      code to execute if condition is false
 ```

````python
if age>18:
    print("You are eligible for voting")
else:
    print("You are a minor")    
````

### **elif statement**
if elif statements are used to check multiple conditions. If the first condition is false, the program will check the next condition, and so on. The else statement is used to execute code if all conditions are false.

 ```python
 if condition:
      code to execute if condition is true
 elif another_condition:
      code to execute if another_condition is true
 else:
      code to execute if all conditions are false
 ```

````python
age = int(input("Enter the age"))
if age<13:
    print("You are a child")
elif age<18:
    print("You are a teenager")
else:
    print("You are an adult")        
````

### **Nested statement**
Nested if statements are if statements that are contained within another if statement. They allow you to check multiple conditions within a single block of code. The syntax for nested if statements is as follows:
 ```python
 if condition:
      code to execute if condition is true
     if another_condition:
          code to execute if another_condition is true
     else:
          code to execute if another_condition is false
 else:
      code to execute if condition is false
```

````python
## number even, odd, negative

num = int(input("Enter the number"))
if num>=0:
    if num == 0:
        print(num,"this number is zero")
        print(num,"this number is even")
    else:    
        print(num,"this number is positive")
        if num%2==0:
            print(num,"this number is even")
        else:
            print(num,"this number is odd")
else:
    print(num,"this number is negative")
    if num % 2 == 0:
        print(num,"this number is even")
    else:
        print(num,"this number is odd")    
````

````python
## Determine if a year is a leap year or not using nested conditional statements.

year = int(input("Enter the year"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year, "is a leap year")
        else:
            print(year,"is not a leap year")    
    else:
        print(year,"is a leap year")        
else:
    print(year,"is not a leap year")
````

### **Complex Example: User Login System**
A simple user login system that checks the username and password.

````python
##User Login System
stored_username = "admin"
stored_password = "password123"
input_username = input("Enter your username:")
input_password = input("Enter your password:")
if input_username == stored_username:
    if input_password == stored_password:
        print("Login successful")
    else:
        print("Incorrect password")
else:
    print("Username not found")    
````

### **Complex Example: Employee Bonus Calculation**
Calculate an employee's bonus based on their performance rating and years of service

````python
## Employee Bonus Calculation
years_of_service = int(input("Enter years of service:"))
performance_rating = float(input("Enter performance rating (1-5):"))
if performance_rating >= 4.5:
    if years_of_service >= 10:
        bonus_percentage = 20
    elif years_of_service >= 5:
        bonus_percentage = 15
    else:
        bonus_percentage = 10
elif performance_rating >= 3.5:
    if years_of_service >= 10:
        bonus_percentage = 15
    elif years_of_service >= 5:
        bonus_percentage = 10
    else:
        bonus_percentage = 5
else:
    bonus_percentage = 0
salary = float(input("Enter your salary:"))
bonus_amount = salary * (bonus_percentage / 100)
print(f"Your bonus is: {bonus_amount:.2f}")    
````
